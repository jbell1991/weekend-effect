```python
# concatenate dataframes
frames = [df_1, df_2, df_3]
df = pd.concat(frames)
df = df.copy()
print(df.shape)
df.tail()
```

```python
# looks like there are some duplicates with same date and timestamp
df = df.drop_duplicates()
print(df.shape)
df.tail()
```

```python
# split date and time into two separate columns
df[['Date','Time']] = df.DateTime.str.split(" ",expand=True)
df = df[['Date', 'Time', 'Open', 'High', 'Low', 'Close']]
df.head()
```

```python
# convert datetime to pandas
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# get day of week to match dates
df['Weekday'] = df['Date'].dt.weekday_name
df = df[['Date', 'Weekday', 'Time', 'Open', 'High', 'Low', 'Close']]
df.head()
```

```python
# the Medallion model began to buy late in the day on a Friday if a clear up-trend existed
# and sell early Monday
# could look at buying at 2:30, 3:00, 3:30 on Friday
# and selling at 9:30, 10:00, 10:30, 11:00 on Monday
```

```python
# selecting Monday mornings and Friday afternoons
friday_afternoon = '15:30:00'
monday_morning = '10:00:00'
weekend = df[((df['Weekday'] == 'Monday') & (df['Time'] == monday_morning)) | 
             ((df['Weekday'] == 'Friday') & (df['Time'] == friday_afternoon)) ]
# removing first row since it's a Monday
weekend = weekend[1:1289]
print(weekend.shape)
weekend.tail(50)
```

```python
# problem - there are more Fridays than Mondays because of market holidays
# when there is a holiday I could put a Tuesday in place of a monday
# just thinking through this there is also Good Friday and holidays that can fall on 
# Friday like Christmas, New Years and the 4th of July
# possible solutions
# I could look only at Friday and Monday pairings with a 3 day difference
# manually go in and fix
weekend['Weekday'].value_counts()
```

```python
weekend['Date'].iloc[1] - weekend['Date'].iloc[0] 
```

```python
# if two mondays or two fridays in a row drop the second row
# we want to get rid of the first friday row
# but the second monday row

# for mondays
weekend_ex_holidays_mon = weekend.groupby((weekend["Weekday"] != weekend["Weekday"].shift()).cumsum().values).first()
weekend_ex_holidays_mon = weekend_ex_holidays_mon[weekend_ex_holidays_mon['Weekday'] == 'Monday']
weekend_ex_holidays_mon
```

```python
# getting correct Fridays
weekend_ex_holidays_fri = weekend.groupby((weekend["Weekday"] != weekend["Weekday"].shift()).cumsum().values).last()
weekend_ex_holidays_fri = weekend_ex_holidays_fri[weekend_ex_holidays_fri['Weekday'] == 'Friday']
weekend_ex_holidays_fri
```

```python
# TODO - concatenate monday and friday dataframes
frames = [weekend_ex_holidays_mon, weekend_ex_holidays_fri]
weekend_ex_holidays = pd.concat(frames)
print(weekend_ex_holidays.shape)
weekend_ex_holidays = weekend_ex_holidays.sort_values(by='Date')
weekend_ex_holidays.head(50)
```

```python
# confirm even number of Mondays and Fridays
weekend_ex_holidays['Weekday'].value_counts()
```

```python
# calculate percentage change
weekend_ex_holidays['Previous_Day_Price'] = weekend_ex_holidays['Open'].shift(1)
weekend_ex_holidays['Percentage_Change'] = ((weekend_ex_holidays['Open'] / 
                                             weekend_ex_holidays['Previous_Day_Price']) - 1) * 100
weekend_ex_holidays
```

```python
# add price direction feature
weekend_ex_holidays['Price_Direction'] = weekend_ex_holidays['Percentage_Change'].apply(
                                         lambda x: 1 if x > 0 else 0)
weekend_ex_holidays
```

```python
# determining if Friday direction
friday_direction = df[((df['Weekday'] == 'Friday') & (df['Time'] == '09:30:00')) | 
                      ((df['Weekday'] == 'Friday') & (df['Time'] == friday_afternoon)) ]
friday_direction.head()
```

```python
# calculate percentage change
friday_direction['Friday_9:30_Price'] = friday_direction['Open'].shift(1)
friday_direction['Percentage_Change_Friday'] = ((friday_direction['Open'] /
                                                 friday_direction['Friday_9:30_Price']) - 1) * 100
friday_direction
```

```python
# add price direction feature
friday_direction['Price_Direction_Friday'] = friday_direction['Percentage_Change_Friday'].apply(
    lambda x: 1 if x > 0 else 0)
# look at Mondays only to get return and direction by holding over the weekend
friday_direction = friday_direction[(friday_direction['Weekday'] == 'Friday') &
                                    (friday_direction['Time'] == friday_afternoon)]
friday_direction = friday_direction[['Date', 'Weekday', 'Friday_9:30_Price','Percentage_Change_Friday', 
                                     'Price_Direction_Friday']]
friday_direction.head()
```

```python
# join Friday returns and directions with weekend returns and directions
weekend_return = pd.merge(weekend_ex_holidays, friday_direction, how='left', on='Date')
weekend_return.head()
```

```python
# add Friday afternoon price
weekend_return['Friday_Afternoon_Price'] = weekend_return['Open']
weekend_return = weekend_return[['Date', 'Weekday_x', 'Time', 'Open', 'High', 'Low', 'Close',
                                 'Previous_Day_Price', 'Percentage_Change', 'Price_Direction',
                                 'Weekday_y', 'Friday_9:30_Price', 'Friday_Afternoon_Price',
                                 'Percentage_Change_Friday', 'Price_Direction_Friday']]
weekend_return.head()
```

```python
# shift Friday data down 1 day to Monday
weekend_return['Weekday_y'] = weekend_return['Weekday_y'].shift(1)
weekend_return['Friday_9:30_Price'] = weekend_return['Friday_9:30_Price'].shift(1)
weekend_return['Friday_Afternoon_Price'] = weekend_return['Friday_Afternoon_Price'].shift(1)
weekend_return['Percentage_Change_Friday'] = weekend_return['Percentage_Change_Friday'].shift(1)
weekend_return['Price_Direction_Friday'] = weekend_return['Price_Direction_Friday'].shift(1)
weekend_return.head()
```

```python
# look at Mondays only to get return and direction by holding over the weekend
weekend_return = weekend_return[weekend_return['Weekday_x'] == 'Monday']
weekend_return.head()
```

```python
# baseline percentage of up and down directional moves over the weekend
weekend_return['Price_Direction'].value_counts(normalize=True)
```

```python
# baseline average return
weekend_return['Percentage_Change'].mean()
```

```python
# create feature of market direction on Friday and Weekend Direction
def label_direction(row):
    if row['Price_Direction'] == 1 and row['Price_Direction_Friday'] == 1:
        return 'Friday Up, Weekend Up'
    if row['Price_Direction'] == 0 and row['Price_Direction_Friday'] == 1:
        return 'Friday Down, Weekend Up'
    if row['Price_Direction'] == 1 and row['Price_Direction_Friday'] == 0:
        return 'Friday Up, Weekend Down'
    if row['Price_Direction'] == 0 and row['Price_Direction_Friday'] == 0:
        return 'Friday Down, Weekend Down'
```

```python
weekend_return['Friday_v_Weekend'] = weekend_return.apply(lambda row: label_direction(row), axis=1)
weekend_return.head(10)
```

```python
percent_change = .75
def label_friday_up_alot(row):
    if row['Percentage_Change_Friday'] >= percent_change and row['Price_Direction'] == 1:
        return f'Friday Up > {percent_change}%, Weekend Up'
    if row['Percentage_Change_Friday'] >= percent_change and row['Price_Direction'] == 0:
        return f'Friday Up > {percent_change}%, Weekend Down'
    if row['Percentage_Change_Friday'] < percent_change and row['Price_Direction'] == 1:
        return f'Friday Not Up > {percent_change}%, Weekend Up'
    if row['Percentage_Change_Friday'] < percent_change and row['Price_Direction'] == 0:
        return f'Friday Not up > {percent_change}%, Weekend Down'
```

```python
weekend_return['Friday_up_alot_v_Weekend'] = weekend_return.apply(lambda row: label_friday_up_alot(row), axis=1)
weekend_return.head(10)
```
